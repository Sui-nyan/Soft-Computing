# Swarm Problem

```python
import random
import numpy as np
```

```python
# Problem 4

def update_hiker(position, velocity, personal_best, global_best):
    #TODO
    # New velocity = 0.7 × old velocity + 0.5 × (personal best - current position) + 0.5 × (global best - current position).
    new_velocity = 0.7 * np.array(velocity) + 0.5 * (np.array(personal_best) - np.array(position)) + 0.5 * (np.array(global_best) - np.array(position))
    # New position = current position + new velocity.
    new_position = np.array(position) + new_velocity

    return new_position, new_velocity

pos = [2,2]
vel = [-0.5, -0.5]
p_best = [1,1]
g_best = [0,0]

update_hiker(pos, vel, p_best, g_best)
```

```python
# Problem 5
# function h(x,y)=x^2+y^2h(x,y)=x^2+y^2
def h(x, y):
    return x**2 + y**2

class Hiker:
    def __init__(self, position, velocity, personal_best, hiker_name=None):
        self.position = position
        self.velocity = velocity
        self.personal_best = personal_best
        self.hiker_name = hiker_name

def random_hiker():
    # Random position in the range [-5, 5]
    position = [random.uniform(-5, 5), random.uniform(-5, 5)]
    # Initial velocity is random in the range [-1, 1]
    velocity = [random.uniform(-1, 1), random.uniform(-1, 1)]
    # Personal best is initially the same as the starting position
    personal_best = position.copy()

    return position, velocity, personal_best

def update_position(position, velocity):
    new_position = np.array(position) + np.array(velocity)
    return new_position.tolist()

def update_personal_best(position, personal_best):
    if h(position[0], position[1]) < h(personal_best[0], personal_best[1]):
        return position
    return personal_best
```

```python
def run_hiker_swarm(num_hikers, num_iterations):
    # Initialize hikers
    hikers = []
    for i in range(num_hikers):
        hikers.append(Hiker(*random_hiker(), hiker_name=f"Hiker_{i}"))

    # Initialize global best
    global_best = min(hikers, key=lambda h: h.personal_best[0]**2 + h.personal_best[1]**2).personal_best

    for _ in range(num_iterations):
        for hiker in hikers:
            # Update velocity and position
            new_position, new_velocity = update_hiker(hiker.position, hiker.velocity, hiker.personal_best, global_best)
            hiker.position = new_position
            hiker.velocity = new_velocity

            # Update personal best
            hiker.personal_best = update_personal_best(hiker.position, hiker.personal_best)
            print(f"{hiker.hiker_name} position: {hiker.position}, personal best: {hiker.personal_best}")

        # Update global best
        global_best = min(hikers, key=lambda h: h.personal_best[0]**2 + h.personal_best[1]**2).personal_best

    return hikers, global_best
```

```python
hikers, global_best = run_hiker_swarm(
    num_hikers=5,
    num_iterations=10
)
```
