````python
import pygame

def setup_gamescreen(width, height, gameloop = None):
    pygame.init()
    screen = pygame.display.set_mode((width, height))

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
        if running:
            screen.fill((0, 0, 0)) # clear the screen

        gameloop(screen) if gameloop else None
        pygame.display.flip() #updates the screen

    pygame.quit()

setup_gamescreen(600, 400)
````

````python
import pygame
import random

width, height = 600, 400
# Snake initialization
snake_block = 10
snake_speed = 15
snake_list = [[width / 2, height / 2]]
snake_direction = "RIGHT"

# Food initialization
food_block = 10
food_x = round(random.randrange(0, width - food_block) / 10.0) * 10.0
food_y = round(random.randrange(0, height - food_block) / 10.0) * 10.0

def get_keypress(press_event, curr_direction):
        if press_event.key == pygame.K_LEFT and curr_direction != "RIGHT":
            return "LEFT"
        elif press_event.key == pygame.K_RIGHT and curr_direction != "LEFT":
            return "RIGHT"
        elif press_event.key == pygame.K_UP and curr_direction != "DOWN":
            return "UP"
        elif press_event.key == pygame.K_DOWN and curr_direction != "UP":
            return "DOWN"

def move_snake(snake_list, direction, snake_block):
    # Snake movement
    snake_head = list(snake_list[-1])
    if direction == "RIGHT":
        snake_head[0] += snake_block
    elif direction == "LEFT":
        snake_head[0] -= snake_block
    elif direction == "UP":
        snake_head[1] -= snake_block
    elif direction == "DOWN":
        snake_head[1] += snake_block

    snake_list.append(snake_head)

def food_collision(snake_list, snake_head, food_x, food_y, width, height, food_block):
    # Check for food collision
    if snake_head[0] == food_x and snake_head[1] == food_y:
        food_x = round(random.randrange(0, width - food_block) / 10.0) * 10.0
        food_y = round(random.randrange(0, height - food_block) / 10.0) * 10.0
    else:
        del snake_list[0]

def gameloop(screen):
    global snake_direction, food_x, food_y
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            return True
        elif event.type == pygame.KEYDOWN:
            new_direction = get_keypress(event, snake_direction)
            if new_direction:
                snake_direction = new_direction

    move_snake(snake_list, snake_direction, snake_block)
    food_collision(snake_list, snake_list[-1], food_x, food_y, width, height, food_block)
    
    screen.fill((0, 0, 0))  # Black background
    pygame.draw.rect(screen, (255, 0, 0), [food_x, food_y, food_block, food_block])
    for segment in snake_list:
        pygame.draw.rect(screen, (0, 255, 0), [segment[0], segment[1], snake_block, snake_block])

    pygame.display.flip()
````

````python
setup_gamescreen(width, height, gameloop)
````

````python
import time

def check_collision(snake_head, snake_list, width, height):
    if snake_head[0] >= width or snake_head[0] < 0 or snake_head[1] >= height or snake_head[1] < 0:
        return True
    for x in snake_list[:-1]:
        if x == snake_head:
            return True
    return False

def game_over(running, screen):
    if not running:
        font_style = pygame.font.SysFont(None, 50)
        message = font_style.render("You Lost!", True, (255, 255, 255))
        screen.blit(message, [width / 3, height / 3])
        pygame.display.flip()
        time.sleep(2) #pause for 2 seconds before quitting.
````
